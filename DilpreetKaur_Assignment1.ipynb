```python
Q1. What are data structures, and why are they important?
Answer:
Data structures are ways of organizing and storing data so it can be accessed and modified efficiently.
They improve performance and make programs easier to manage.Eg-List,Tuple,dictionary,set.
Importance of data structures:
1. Efficient Data Storage
Data structures allow large amounts of data to be stored in an organized way.
2. Faster Data Retrieval
Some data structures allow very fast searching and retrieval of data.Eg-Dictionary
3. Better Memory Management
Proper data structures help reduce memory usage and improve program performance.
For example, a tuple uses less memory than a list because tuples are immutable.
```

```python
Q2. Explain the difference between mutable and immutable data types with examples.
Ans:
Mutable Data Types:Mutable objects are those whose contents can be changed after they are created.
Key Points:
Their values can be modified after creation.
The same object is updated in memory.
They support operations like add, delete, and update.
Generally cannot be used as dictionary keys.
Useful when data needs frequent modifications.
Examples of Mutable Data Types
List (list)
Dictionary (dict)
Set (set)
Example
numbers = [1, 2, 3]
numbers.append(4)
print(numbers)
Output:
[1, 2, 3, 4]-The original list is modified.

Immutable Data Types:
Immutable objects are those whose contents cannot be changed after they are created.
Key Points:
Their values cannot be modified after creation.
Any change creates a new object.
They are generally more memory-efficient.
They can be used as dictionary keys (if hashable).
They provide better data security and integrity because they cannot be modified accidentally, the data remains reliable.
Examples of Immutable Data Types
String (str)
Tuple (tuple)
Integer (int)
Float (float)
Boolean (bool)
Example
name = "python"
name = name.upper()
print(name)
Output:
PYTHON-A new string object is created because strings are immutable.
```

```python
Q3. What are the main differences between lists and tuples in Python?
ans:
List
A list is an ordered collection of elements that can be modified after creation.
Characteristics of Lists
Lists are mutable.
Elements can be added, removed, or updated.
Lists are created using square brackets [].
Lists have many built-in methods such as append(), remove(), and insert().
Lists use more memory compared to tuples.
Tuple
A tuple is an ordered collection of elements that cannot be modified after creation.
Characteristics of Tuples
Tuples are immutable.
Elements cannot be added, removed, or updated.
Tuples are created using parentheses ().
Tuples have fewer built-in methods.
Tuples use less memory and are generally faster than lists.
```

```python
Q4. Describe how dictionaries store data.
ans:
A dictionary is a data structure that stores data in the form of key-value pairs.
Each key is unique and is associated with a specific value.dictionary internally uses a hash table.
When a key is provided, Python calculates its hash value and directly locates the associated value, making dictionary operations very efficient and fast.
eg:
student = { "name": "dil",  "age": 20, "course": "Python"}
```

```python
Q5. Why might you use a set instead of a list in Python?
ans:
A set is used instead of a list when we want unique values and faster searching.
Key Points
Sets automatically remove duplicate values.
Searching is usually faster in a set.When you search in a set, Python uses the element's hash value to jump directly to its location.
Useful when only unique data is needed.
Easy to remove duplicates from a list.
Supports operations like union and intersection.
```

```python
Q7. How do tuples ensure data integrity in Python?
ans:
Data integrity means protecting data from accidental or unauthorized changes.
A tuple ensures data integrity because it is immutable, meaning its elements cannot be changed after creation.
Key Points
Tuples are immutable.
Elements cannot be added, removed, or modified.
This prevents accidental changes to important data.
Data remains consistent throughout the program.
Therefore, tuples help maintain data integrity.
```

```python
Q8. What is a hash table, and how does it relate to dictionaries in Python?
A hash table is a data structure that stores data in such a way that values can be found very quickly.
It uses a hash function to convert a key into a unique number called a hash value, which helps determine where the data should be stored.
Python dictionaries are implemented using hash tables.
When a key is provided, Python calculates its hash value and directly locates the associated value, making dictionary operations very efficient and fast
```

```python
Q10. Explain why strings are immutable in Python.
ans:
A string is immutable, which means its characters cannot be changed after the string is created.
1. Better Data Security
Strings often store important data such as passwords, file paths, and URLs. Immutability prevents accidental changes.

2. Memory Efficiency
Python can reuse the same string object in memory when multiple variables have the same value.

3. Faster Dictionary Operations
Since strings never change, their hash value remains constant, making dictionary lookups fast and reliable.

4. Data Integrity
The value of a string cannot be changed accidentally, so the data remains consistent.
```

```python
Q17. Describe the time complexity of accessing elements in a dictionary.
ans:
The average time complexity of accessing an element in a dictionary is O(1), or constant time. 
This is because dictionaries use a hash table, which allows Python to directly locate a value using its key instead of searching through all elements one by one.
Therefore, dictionary lookups are usually very fast regardless of the dictionary's size.
```

```python
Q19. Why are dictionaries considered unordered, and how does that affect data retrieval?
ans:
Dictionaries were traditionally considered unordered because they store data using a hash table rather than index positions. This allows Python to retrieve values quickly using keys.
In modern Python (3.7+), dictionaries preserve insertion order, but data is still accessed through keys rather than indexes.
```
